```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('nba_final.csv')
```

```python
data['Pos2'] = data['Pos2'].replace(np.nan,'NONE')
data['Pos2'].unique()
```

```python
data.Pos2.value_counts()
```

```python
data.isna().sum()
```

```python
data = data.fillna(0)
```

```python
data.isna().sum()
```

```python
data.head(10).T
```

```python
data = data.drop(['Player.x','Player_ID'],axis = 1)
```

```python
# to deal with categorical variables

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Pos1'] = le.fit_transform(data['Pos1'])

data['Pos2'] = le.fit_transform(data['Pos2'])

data['Tm'] = le.fit_transform(data['Tm'])

data['Season'] = le.fit_transform(data['Season'])

data['Conference'] = le.fit_transform(data['Conference'])

data['Role'] = le.fit_transform(data['Role'])

data['Play'] = le.fit_transform(data['Play'])
```

```python
data.Play.value_counts(normalize= True)
```

```python
X= data.drop(['Play'],axis = 1)
y = data['Play']
```

```python
# standardize 

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(X)
```

```python
x
```

```python
# LDA

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=1)
x = lda.fit_transform(x,y)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state = 29)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train,y_train)
lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report

cr = classification_report(lr_pred,y_test)
print(cr)
```

```python
#decision tree classifier

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
dt_pred = dt.predict(x_test)
```

```python
cr_dt = classification_report(dt_pred,y_test)
```

```python
print(cr_dt)
```

```python
#random forest classifier

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
pred_rf = rf.predict(x_test)
```

```python
cr_rf = classification_report(pred_rf,y_test)
```

```python
print(cr_rf)
```
